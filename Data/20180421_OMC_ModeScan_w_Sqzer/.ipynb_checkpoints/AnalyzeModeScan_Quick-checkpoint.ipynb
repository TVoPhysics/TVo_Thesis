```python
data_center = np.loadtxt('20180422_SQZ_OMC_Scan_OMC_ASC_ON_Center.txt')
data_close = np.loadtxt('20180422_SQZ_OMC_Scan_OMC_ASC_ON_Closest.txt')
data_furth = np.loadtxt('20180422_SQZ_OMC_Scan_OMC_ASC_ON_Furthest.txt')
```

```python
import numpy as np
from scipy.signal import butter, lfilter, freqz
import matplotlib.pyplot as plt


def butter_lowpass(cutoff, fs, order=5):
    nyq = 0.5 * fs
    normal_cutoff = cutoff / nyq
    b, a = butter(order, normal_cutoff, btype='low', analog=False)
    return b, a

def butter_lowpass_filter(data, cutoff, fs, order=5):
    b, a = butter_lowpass(cutoff, fs, order=order)
    y = lfilter(b, a, data)
    return y

# Filter requirements.
order = 6
fs = 16384       # sample rate, Hz
cutoff = 500  # desired cutoff frequency of the filter, Hz

# Get the filter coefficients so we can check its frequency response.
b, a = butter_lowpass(cutoff, fs, order)

# Plot the frequency response.
w, h = freqz(b, a, worN=8000)
plt.subplot(2, 1, 1)
plt.loglog(0.5*fs*w/np.pi, np.abs(h), 'b')
plt.loglog(cutoff, 0.5*np.sqrt(2), 'ko')
plt.axvline(cutoff, color='k')
plt.xlim(0, 0.5*fs)
plt.title("Lowpass Filter Frequency Response")
plt.xlabel('Frequency [Hz]')
plt.grid()
```

```python
def applyFilt_plot(data, cutoff, fs, order, guess_peak_00, guess_peak_20 , title):
    # Filter the data, and plot both the original and filtered signals.
    y = butter_lowpass_filter(data[:,2], cutoff, fs, order)
    
    #guess the time of the peaks for 02 and 00
    sec1=round(guess_peak_00[0]*fs/8)
    sec2=round(guess_peak_00[1]*fs/8)
    peakind00=scipy.signal.find_peaks_cwt(y[sec1:sec2],np.arange(sec1/5000,sec2/5000))
    #print(peakind00+sec1,y[peakind00+sec1])
    #print(data[peakind00+sec1,0])
    
    
    sec1_20=round(guess_peak_20[0]*fs/8)
    sec2_20=round(guess_peak_20[1]*fs/8)
    #print(sec1,sec2)
    peakind20=scipy.signal.find_peaks_cwt(y[sec1_20:sec2_20],np.arange(sec1_20/5000,sec2_20/5000))
    #print(peakind20+sec1_20,y[peakind20+sec1_20])
    #print(data[peakind20+sec1_20,0])
    
    print("TEM02 Peak is "+str(max(y[peakind20+sec1_20])))
    print("TEM00 Peak is "+str((max(y[peakind00+sec1]))))
    print("Mismatch Estimate is "+ str((max(y[peakind20+sec1_20])/max(y[peakind00+sec1]))))
    
    
    plt.figure(figsize=(12,7))
    
    plt.subplot(2, 1, 1)
    plt.title(title)
    plt.plot(data[:,0],data[:,1])
    plt.xlabel('Time [sec]')
    plt.ylabel('PZT [Volts]')
    plt.xlim( data[round(sec1-(5*fs/8)),0],data[round(sec1+(10*fs/8)),0] )
    plt.grid()
    
    plt.subplot(2, 1, 2)
    plt.semilogy(data[:,0], data[:,2], 'b-', label='Raw data')
    plt.semilogy(data[:,0], y, 'g-', linewidth=2, label='Filtered data')
    plt.plot(data[peakind00+sec1,0],y[peakind00+sec1],'ro', label='TEM00 Peaks')
    plt.plot(data[peakind20+sec1_20,0],y[peakind20+sec1_20],'mo', label='TEM02 Peaks')
    plt.xlabel('Time [sec]')
    plt.ylabel('OMC_DCPD [mW]')
    plt.xlim( data[round(sec1-(5*fs/8)),0],data[round(sec1+(10*fs/8)),0] )
    plt.ylim(1e-4,1e-1)
    plt.grid()
    plt.legend()
    plt.subplot(2, 1, 1)
    
    plt.show()
    plt.savefig(title+'.png')
    return y
```

```python
guess_peak_20=[6.5,7.5]
guess_peak_00=[10.,11.]
applyFilt_plot(data_center, cutoff, fs, order,guess_peak_00,guess_peak_20,'Lens_in_Center_Position')
```
