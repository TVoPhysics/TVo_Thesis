```python
import matplotlib.pyplot as pl
import numpy as np

%matplotlib inline
```

```python
def sinewave(A_in, dA, p_in, dp, deg):
    A, sigma_a = A_in, dA
    dA = np.random.normal(A, sigma_a, len(deg) )

    phi, sigma_p = p_in, dp 
    dphi = np.random.normal(phi, sigma_p, len(deg) )

    f = (A+dA)*np.sin(deg+phi+dphi)
    return f
```
