```python
import pykat
import matplotlib.pyplot as pl
import numpy as np

%matplotlib inline
```

```python
pykat.BeamParam??
```

```python
w0 = 1e-3

z = np.arange(-10.0,10.0,0.1)
q = pykat.BeamParam(w0=w0, z=z)

zr = q.zr[0]
```

```python
n = np.arange(-2,3,1)
nzr = [n*zr]

nzr_labels = ['-2$z_0$','-$z_0$','0','$z_0$','2$z_0$']
```

```python
fig = pl.figure(dpi=300)

ax = fig.add_subplot(311)
ax.fill_between(z, -q.w, q.w)
ax.set_xticklabels([])

pl.ylabel("Beam Size [m]")
pl.ticklabel_format(style='sci', axis='y', scilimits=(0,0))
pl.grid()

for xc in nzr[0]:
    pl.axvline(x=xc, linestyle='--',color='grey')

ax2 = ax.twiny()

ax2.set_xlim(ax.get_xlim())

ax2.set_xticks(nzr[0])
ax2.set_xticklabels(nzr_labels)

ax2.set_xlabel("Rayleigh Ranges")


bx = fig.add_subplot(312)
bx.plot(z,q.Rc)

bx.set_xticklabels([])

pl.ylim(-30,30)
pl.ylabel("ROC [m]")

pl.grid()

cx = fig.add_subplot(313)
cx.plot(z, (180./np.pi)*q.gouy())

pl.ylabel("$\phi(z)$ [Deg]")

pl.xlabel("Distance from Waist [m]")
pl.grid()

pl.savefig("../../Figures/GaussBeamParams.png")
```
