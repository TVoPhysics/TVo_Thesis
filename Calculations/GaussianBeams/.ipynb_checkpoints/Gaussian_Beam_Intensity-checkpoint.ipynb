```python
w0 = 1e-3

q = pykat.BeamParam(w0=w0, z=z)

zr = q.zr
```

```python
r = np.arange(-1e-2,1e-2,1e-4)

def get_i(w):
    intensity = (q.w0 / w)**2 * np.exp(-2.*r**2/(w**2.))
    return intensity

def get_w(z):
    w = w0 * (1 + (z/zr)**2 )**(0.5)
    return w

wz0 = get_w(0)
wz1 = get_w(zr)
wz2 = get_w(2*zr)

intensity0 = get_i(wz0)
intensity1 = get_i(wz1)
intensity2 = get_i(wz2)
```

```python
fig = pl.figure(dpi=300)

ax = fig.add_subplot(111)
ax.plot(r,intensity0,label='z=0')
ax.plot(r,intensity1,label='$z=z_R$')
ax.plot(r,intensity2,label='$z=2z_R$')

pl.xlabel("Radius [m]")
pl.ylabel("Normalized Intensity [W/m]")

pl.ticklabel_format(style='sci', axis='x', scilimits=(0,0))
pl.legend()

pl.grid()

pl.savefig("../../Figures/GaussBeamIntensity.png")
```
