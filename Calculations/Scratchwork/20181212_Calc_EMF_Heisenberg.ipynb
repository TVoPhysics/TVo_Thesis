```python
import matplotlib.pyplot as pl
import numpy as np

%matplotlib inline
```

```python
def sinewave(A_in, dA, p_in, dp, deg):
    A, sigma_a = A_in, dA
    dA = np.random.normal(A, sigma_a, len(deg) )

    phi, sigma_p = p_in, dp 
    dphi = np.random.normal(phi, sigma_p, len(deg) )

    f = (A+dA)*np.sin(deg+phi+dphi)
    return f
```

```python
n=5000
deg = np.linspace(0,20,n)
```

```python
n
```

```python
dp_in = np.sqrt(0.1)
dA_in=0.1*(1/dp_in)

print(dp_in,dA_in)
sinewave_ar_base = sinewave(1.0, dA_in, 0.0, dp_in, deg)


dp_in = np.sqrt(0.001)
dA_in=0.1*(1/dp_in)

print(dp_in,dA_in)

sinewave_ar_phase_sqz = sinewave(1.0, dA_in, 0.0, dp_in, deg)
```

```python
############################ Plot
pl.figure(figsize=(15, 8), dpi=80)

############################
pl.subplot(111)

pl.plot(deg,sinewave_ar_base,'o')
pl.plot(deg,sinewave_ar_phase_sqz,'o')

pl.xlabel("Degrees",fontsize=18)
pl.ylabel("Electric Field",fontsize=18)
pl.tick_params(labelsize=16)
pl.grid()
pl.legend(prop={'size': 16})
```
