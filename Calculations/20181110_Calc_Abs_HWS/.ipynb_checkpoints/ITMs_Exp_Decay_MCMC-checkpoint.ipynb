```python
%matplotlib inline
from __future__ import division
import numpy as np
np.random.seed(123456789)   # repeatable results
import matplotlib.pyplot as pl
import matplotlib.mlab as mlab
import scipy.signal as sig
import scipy.constants as const
from scipy import interpolate
import corner
import emcee as mc
import pickle
```

```python
### Load comsol model for thermal lens
# Comsol model of the self heating
model = np.loadtxt('1W_self_heating_defocus_doublepass.txt')

time_model = model[:,0]
abs_model  = model[:,1]

def TL(x,a,b):    
    y = interpolate.interp1d(time_model,-abs_model,fill_value="extrapolate")
    ynew = y(x)
    return a + (b * ynew)

a=0.0
b=1.0
x = np.arange(0,1,.001) #hours
f = TL(x,a,b)

#### Plot
pl.figure(figsize=(5, 3), dpi=80)
pl.plot(x,f)
pl.plot(time_model,-abs_model, 'o')
pl.grid()
pl.show()
```

```python
### Best guess for ITM
### Test on ITMX numberst
amp_true  =  -5.85791125 #Starting Amplitude Guess
tau_true = 5.02301549e-02 # Decay Time guess

print('amp_model = ', amp_true)
print('tau_model = ', tau_true)
```

```python
### Load the HWS Data
itmx = 'data_ITMX.p'
itmy = 'data_ITMY.p'

i_ind_x = 80
e_ind_x = -1

i_ind_y = 40
e_ind_y = -1

def open_file(filename):
    with open(str(filename), 'rb') as f:
        u = pickle._Unpickler(f)
        u.encoding = 'latin1'
        data_p = u.load()
    return data_p

itmx_data_p = open_file(itmx)
itmy_data_p = open_file(itmy)

itmx_data = itmx_data_p[(700,450)]['spherical']
t_vect_x = np.arange(0.,float(len(itmx_data)),1.)

t_vect_x = (t_vect_x[ i_ind_x:e_ind_x ])*20/3600
itmx_data = itmx_data[ i_ind_x:e_ind_x ]*1e6


itmy_data = itmy_data_p[(700,560)]['spherical'][ i_ind_y:e_ind_y ]

t_vect_x = np.arange(0.,float(len(itmx_data)),1.)
t_vect_x = (t_vect_x)*20/3600
t_vect_x = t_vect_x - t_vect_x[0]
itmx_data = itmx_data-itmx_data[0]


t_vect_y = np.arange(0.,float(len(itmy_data)),1.)
t_vect_y = (t_vect_y*20.)

signal = itmx_data

pl.plot(t_vect_x, itmx_data, 'o')
pl.grid()
```

```python
# Define Log Likelihood

noise_std = 30.0
uncs = noise_std * np.ones(len(t_vect_x))  # Uniform uncertainty in each point

def lnprob(theta, t, signal, uncs) :
    amplitude = theta[0]
    tau = theta[1]
    
    if amplitude <= -30.0 or amplitude > 30.0 or a <= -1.0 or a > 1.0 :
        return -np.inf
    else :
        new_signal = TL(t,amplitude,tau)
        return -np.sum(np.abs(signal - new_signal)**2/(2*np.square(uncs)))
```

```python
ndim = 2
nwalkers = 100
rr = np.transpose(np.vstack([np.random.normal(-1.0, 1.0, nwalkers),
                             np.random.normal(0, 5.0, nwalkers) ]))
p0 = np.tile(np.array([amp_true, tau_true]), (nwalkers,1)) + rr
rr[0]
```

```python
samp = mc.EnsembleSampler(nwalkers, ndim, lnprob, args=[t_vect_x, signal, uncs])
```

```python
pos, prob, state = samp.run_mcmc(p0, 1000)
samp.reset()
```

```python
samp.run_mcmc(pos, 5000);
```

```python
results = np.copy(samp.flatchain) 
```

```python
range_list = [ [-30, 30], [0.0, 0.1] ]
h_c = corner.corner(results, 
                    bins=50, 
                    labels=['Amplitude', 'Decay'],
                    show_titles=True,
                    title_kwargs={'fontsize':14},
                    title_fmt='.3f' ,
                    quantiles=[0.16, 0.5, 0.84],
                    range=range_list,
                    verbose=True,
                    hist_kwargs={'color':'b'})
h_c.savefig("SineGaussianTrianglePlots.pdf")
```

```python
tc_mean, amp_mean, a_mean = np.mean(results,axis=0)
tc_median, amp_median, a_median = np.median(results,axis=0)
covmat = np.cov(results.T)
tc_unc, amp_unc, a_unc = np.sqrt(np.diag(covmat))
print("Means         = ", tc_mean, amp_mean, a_mean)
print("Medians       = ", tc_median, amp_median, a_median)
print("Uncertainties = ", tc_unc, amp_unc, a_unc)
print("Truths        = ", tc_true, amp_true, tau_true)
print(covmat)

recovered_exp = TL(t_vect_x,amp_median,a_median)
pl.plot(t_vect_x, np.real(recovered_exp), '--',label='Recovered exponential function')
pl.plot(t_vect_x, signal, 'o', label='Noisy exponential function')
pl.legend(loc='best');
pl.show();
#pl.plot(t_vect_x, np.real(exp_c), label='Actual exponential function')
pl.plot(t_vect_x, np.real(recovered_exp), '--', label='Recovered exponential function')
pl.legend(loc='best');
pl.xlabel('Time (s)')
pl.ylabel('Amplitude')
pl.title('Input exponential function vs. MCMC Recovered exponential function')
pl.savefig('ActualvsRecoveredExp.pdf')
pl.show();
#pl.plot(t_vect_x, np.real(recovered_exp) - np.real(exp_c), label='Residual (Recovered - Actual)')
pl.legend(loc='best');
```
