```python
def cav2node(cav_name, node, verbose =True, base=base):
    """
    This method traces a cavity's eigen mode to a
    given node in the model. 
    
    inputs
    cav_name: cavity name
    node:     node to trace beam to
    verbose:  More or less info
    base:     The kat object to do the tracing with
    
    returns
    qx, qy:   q-parameters in the two transverse directions
    """
    kat = base.deepcopy()

    # turn off all cav commands except the one you want.
    for cav in kat.getAll(pykat.commands.cavity):
        if cav_name not in cav.name:
            cav.remove()

    kat.noxaxis = True
    kat.parse("pd p nAS")

    # run the kat file to get trace data
    _, T = kat.run(getTraceData=True)
    qx, qy, _ = T[0][node]
    
    if verbose: print("%s -> %s" % (cav_name, node), qx.q, qy.q)
    return qx, qy
```

```python
def single_bounce_SQZ(base, direction = 'x', verbose=True, to_node='nAS'):
    """
    This does a single bounce trace off of SRM HR surface from the OPO beam.
    This happens in two parts. Once trace from the OPO to the SRM HR
    a refl off of SRM HR, and then a trace from this reflected beam to
    the requested node.
    
    inputs==
    base:          The kat object to do the tracing with
    direction:     Transverse direction, turns out with astimagitism, this is important
    verbose:       True, print
    node:          nAS, gotta choose somewhere to measure the q-reference
    
    returns==
    the final q value going in to the OMC
    """
    
    SQZqx = base.sqz.nSQZ.qx 
    SQZqy = base.sqz.nSQZ.qy
    
    if direction == 'y':
        qin = SQZqy
    else:
        qin = SQZqx

    #trace beam from squezer to SRM
    SQZ2SRM = base.beamTrace(qin, "nSQZ", "nSRMs1", direction=direction)
    
    #ITMX substrate
    n = base.SRMsub.n.value

    # Manually calculate the reflection off of SRM HR inside substrate
    q2 = pykat.optics.ABCD.apply(base.SRM.ABCD("nSRMs1", "nSRMs1", direction=direction),
                                 SQZ2SRM.q_out,
                                 n, n)

    if to_node != "nSRMs1":
        SRM2OMC = base.beamTrace(q2, "nSRMs1", to_node, direction=direction)

        if verbose:
            SQZ2SRM.print()
            SRM2OMC.print()

        return pykat.BeamParam(q=SRM2OMC.q_out)
    else:
        if verbose:
            SQZ2SRM.print()
        
        return pykat.BeamParam(q=SQZ2SRM.q_out)
```

```python
def shift_lenses(kat, lens1_dz, lens2_pos, lens2_dz=0):
        """
        Shifts lenses relative to those that have been measured in H1 OPO->ZM1 path
        
        inputs ==
        lens1_dz - change in lens position in m, positive closer to OPO block
        lens2_pos - 0 to 1 value for position along translation stage
        lens2_dz - change in lens position in m, positive closer to OPO block
        
        returns ==
        none
        """
        
        lens2_dz += (2*lens2_pos-1) * kat.data['lens2_range']
        
        kat.lsqz_lens1.L         = kat.data['OPO_block_to_lens1'] - lens1_dz
        kat.lsqz_lens1_faraday.L = kat.data['OPO_lens1_to_faraday'] + lens1_dz
        kat.lsqz_faraday_lens2.L = kat.data['OPO_faraday_to_lens2'] - lens2_dz
        kat.lsqz_lens2_zm1.L     = kat.data['OPO_block_to_ZM1'] - kat.lsqz_lens1.L.value - kat.lsqz_lens1_faraday.L.value - kat.lsqz_faraday_lens2.L.value
        
        if kat.lsqz_lens1.L < 0:         raise Exception("Uh oh")
        if kat.lsqz_lens1_faraday.L < 0: raise Exception("Uh oh")
        if kat.lsqz_faraday_lens2.L < 0: raise Exception("Uh oh")
        if kat.lsqz_lens2_zm1.L < 0:     raise Exception("Uh oh")
```

```python
def shift_lenses1(in_kat, lens1_dz, lens2_pos, lens2_dz=0):
        """
        Shifts lenses relative to those that have been measured in H1 OPO->ZM1 path
        actually remakes the kat object, easier to use once we've defined a good base.
        
        inputs ==
        lens1_dz - change in lens position in m, positive closer to OPO block
        lens2_pos - 0 to 1 value for position along translation stage
        lens2_dz - change in lens position in m, positive closer to OPO block
        
        returns ==
        kat object
        """
        
        out_kat = in_kat.deepcopy()
        
        lens2_dz += (2*lens2_pos-1) * in_kat.data['lens2_range']
        
        out_kat.lsqz_lens1.L         = in_kat.lsqz_lens1.L   - lens1_dz
        out_kat.lsqz_lens1_faraday.L = in_kat.lsqz_lens1_faraday.L + lens1_dz
        out_kat.lsqz_faraday_lens2.L = in_kat.lsqz_faraday_lens2.L- lens2_dz
        out_kat.lsqz_lens2_zm1.L     = in_kat.lsqz_lens2_zm1.L  + lens2_dz
        
        if out_kat.lsqz_lens1.L < 0:         raise Exception("Uh oh")
        if out_kat.lsqz_lens1_faraday.L < 0: raise Exception("Uh oh")
        if out_kat.lsqz_faraday_lens2.L < 0: raise Exception("Uh oh")
        if out_kat.lsqz_lens2_zm1.L < 0:     raise Exception("Uh oh")
            
        return out_kat
```

```python
def func_lens2_fit(x, dz1, dz2, direction, base):
    '''
    Seems like this guy shifts lens2 to fit the data.
    
    '''

    kat = base.deepcopy()

    lens2_actuator_pos = 0.5
    
    if direction == 'x':
        qin = kat.sqz.nSQZ.qx
    else:
        qin = kat.sqz.nSQZ.qy
    
    kat.data['fn_OPO_shift_lenses'](kat, dz1, lens2_actuator_pos, dz2)
    
    trace_SQZ_SRM = kat.beamTrace(qin, 'nSQZ', "nSRMs1", direction=direction)

    return trace_SQZ_SRM.beamsize(x)/1e-3
```

```python
def func_lens1_fit(x, dz1, direction, base):
    '''
    Seems like this guy shifts lens1 to fit the data.
    
    '''
    kat = base.deepcopy()

    lens2_actuator_pos = 0.5
    
    if direction == 'x':
        qin = kat.sqz.nSQZ.qx
    else:
        qin = kat.sqz.nSQZ.qy
    
    kat.sqz_lens2.f = np.inf
    
    kat.data['fn_OPO_shift_lenses'](kat, dz1, lens2_actuator_pos, 0)
    
    trace_SQZ_SRM = kat.beamTrace(qin, 'nSQZ', "nSRMs1", direction=direction)
    
    return trace_SQZ_SRM.beamsize(x)/1e-3
```

```python
def func_lens1_fitx(x, dz1):
    return func_lens1_fit(x, dz1, 'x', kat)

def func_lens1_fity(x, dz1):
    return func_lens1_fit(x, dz1, 'y', kat)

def func_lens2_fitx(x, dz2):
    return func_lens2_fit(x, dz1, dz2, 'x', kat)

def func_lens2_fity(x, dz2):
    return func_lens2_fit(x, dz1, dz2, 'y', kat)
```

```python
def fit_input_q(qguess, z_data, w_data, eps_data, z_ref=None):
    from lmfit import minimize, Parameters
    
    qguess = pykat.BeamParam(q=complex(qguess))
    
    def residual(params, kat, z, data, eps_data):
        q0 = pykat.BeamParam(z=params['z0'], w0=params['w0'])
        T = kat.beamTrace(q0, 'nSRMs1', 'nOMC_ICa')
        model = T.beamsize(z, z_ref=z_ref)

        return (data-model) / eps_data

    params = Parameters()
    params.add('w0', value=qguess.w0, min=0)
    params.add('z0', value=qguess.z)

    out = minimize(residual, params, args=(kat, z_data, w_data, eps_data))

    qFIT = pykat.BeamParam(z=out.params['z0'], w0=out.params['w0'])

    return qFIT, out
```

```python
def fit_input_q_OPO(qguess, z_data, w_data, eps_data, z_ref=None):
    from lmfit import minimize, Parameters
    
    qguess = pykat.BeamParam(q=complex(qguess))
    
    def residual(params, kat, z, data, eps_data):
        q0 = pykat.BeamParam(z=params['z0'], w0=params['w0'])
        T = kat.beamTrace(q0, kat.ZM1.nZM1a, 'nSRM2')
        model = T.beamsize(z, z_ref=z_ref)

        return (data-model) / eps_data

    params = Parameters()
    params.add('w0', value=qguess.w0, min=0)
    params.add('z0', value=qguess.z)

    out = minimize(residual, params, args=(kat, z_data, w_data, eps_data))

    qFIT = pykat.BeamParam(z=out.params['z0'], w0=out.params['w0'])

    return qFIT, out
```

```python
### Trace the beam without lens2 to find the lens 1 positioning, two transverse directions.

dz1 = 0
dz2 = 0

kat = base.deepcopy()

#Trace the squeezer bame to the SRM HR node and get the ZM1 position
trace_SQZ_SRM_HR = kat.beamTrace(base.sqz.nSQZ.qx, 'nSQZ', "nSRMs1", direction='x')
ZM1_pos          = trace_SQZ_SRM_HR.data['ZM1']['z']

# xdata is the propagation axis with the input/output coupler of the OPO as x=0
# ydata is the beam radius

xdata = ZM1_pos + OPO_wo_lens2[0,:]
ydata = OPO_wo_lens2[1,:]/2e-3

# use the least squares fitting to find the right lens1 positioning
p0x = (0)
popt_lens1x, covx = curve_fit(func_lens1_fitx, xdata, ydata, p0=p0x)

# xdata is the propagation axis with the input/output coupler of the OPO as x=0
# ydata is the beam radius
xdata = ZM1_pos + OPO_wo_lens2[0,:]
ydata = OPO_wo_lens2[2,:]/2e-3

# use the least squares fitting to find the right lens1 positioning
p0y = (0)
popt_lens1y, covy = curve_fit(func_lens1_fity, xdata, ydata, p0=p0y)

# use the average of the two transverse directions to get the approximate distance from OPO to lens1
dz1 = (popt_lens1x + popt_lens1y)/2

print("Fitted distance from OPO to L1 " + str(dz1[0]))
```

```python
def SRM_HR_substrate_refl(qix, qiy):
    qx = pykat.optics.ABCD.apply(base.SRM.ABCD("nSRMs1", "nSRMs1", direction='x'),
                                     qix, kat.SRMsub.n.value, kat.SRMsub.n.value)
    qy = pykat.optics.ABCD.apply(base.SRM.ABCD("nSRMs1", "nSRMs1", direction='y'),
                                     qiy, kat.SRMsub.n.value, kat.SRMsub.n.value)
    
    return qx, qy
```

```python
def run(base, dz1=dz1, dz2=dz2,fname=None,lens_config=None):
    """
    This function runs the provided kat object to trace the beam form the
    OPO to the OMC. It plots the data taken against the model generated.
    The parameters of the model can be changed by changing the input kat.
    The popt x/y are the optimised qx and qy from the fitting before of the
    beam coming out of the OPO after ZM1.
    """
    fig = plt.figure(figsize=(8, 8),dpi=300)
    ax = plt.subplot(211)
    ax2 = plt.subplot(212)

    # Get cavity OMC q at SRM...
    OMCqx, OMCqy = cav2node('OMC', 'nSRMs1', base)
    OMCqx.reverse()
    OMCqy.reverse()
    
    # Then trace it back to get the beam shape
    trace_OMCx = base.beamTrace(OMCqx, 'nSRMs1', "nOMC_REFL",direction='x')
    trace_OMCx.plot_beamsize(ax=ax2, color='k', ls='--', label="OMCx")
    trace_OMCy = base.beamTrace(OMCqy, 'nSRMs1', "nOMC_REFL",direction='y')
    trace_OMCy.plot_beamsize(ax=ax2, color='k', ls=':', label="OMCy")
    
    
    ### Plot fitting from moving lens2
    
    # Loop over the three lens2 options and do the tracing
    for lens2_pos,c,pos in zip([0.0, 0.5, 1.0], ['r','g','b'], ['Far', 'Middle', 'Close']):
        
        print(lens2_pos,c,pos)
        
        kat = shift_lenses1(base, 0, lens2_pos, 0)
        
        trace_SQZ_SRMx = kat.beamTrace(kat.sqz.nSQZ.qx, 'nSQZ', "nSRMs1",direction='x')
        trace_SQZ_SRMx.plot_beamsize(ax=ax, color=c, label=pos + '(x)')
        trace_SQZ_SRMy = kat.beamTrace(kat.sqz.nSQZ.qy, 'nSQZ', "nSRMs1",direction='y')
        trace_SQZ_SRMy.plot_beamsize(ax=ax, color=c, label=pos + '(y)', ls='--')
        
        qx, qy = SRM_HR_substrate_refl(trace_SQZ_SRMx.q_out, trace_SQZ_SRMy.q_out)
        
        trace_SRM_OMCx = kat.beamTrace(qx, 'nSRMs1', "nOMC_REFL",direction='x')
        trace_SRM_OMCx.plot_beamsize(ax=ax2, color=c, label=pos + '(x)')
        trace_SRM_OMCy = kat.beamTrace(qy, 'nSRMs1', "nOMC_REFL",direction='y')
        trace_SRM_OMCy.plot_beamsize(ax=ax2, color=c, label=pos + '(y)', ls='--')

        print(pos, "x", pykat.BeamParam.overlap(trace_SRM_OMCx.q_out, trace_OMCx.q_out))
        print(pos, "y", pykat.BeamParam.overlap(trace_SRM_OMCy.q_out, trace_OMCy.q_out))
        print(kat.lsqz_lens1.L , kat.lsqz_lens1_faraday.L , kat.lsqz_faraday.L , kat.lsqz_faraday_lens2.L)

        print('Distance from OPO to lens2 is '+
              str(kat.lsqz_lens1.L+kat.lsqz_lens1_faraday.L+kat.lsqz_faraday.L+kat.lsqz_faraday_lens2.L)+ ' [m]')

    ZM1_pos = trace_SQZ_SRM_HR.data['ZM1']['z']
    OM1_pos = trace_OMCx.data['OM1']['z']
    if lens_config is 'Before':
        ### Plot measured data
        OMC_waist_pos = trace_OMCx.data['OM3']['z'] + 0.117 + 0.2 + 0.141
        ZM1_pos = trace_SQZ_SRM_HR.data['ZM1']['z']

        # Factor of two because measurement in diameter
        ax.scatter(ZM1_pos + OPO[0,:], OPO[1,:]/2e-3, color='g', s=80, marker='x')
        ax.scatter(ZM1_pos + OPO[0,:], OPO[2,:]/2e-3, color='g', s=80, marker='+')

        ax2.scatter(OMC_waist_pos + Furthest[0,:], Furthest[1,:]/1e-3, color='r', s=80, marker='x')
        ax2.scatter(OMC_waist_pos + Middle[0,:],   Middle[1,:]/1e-3, color='g', s=80, marker='x' )
        ax2.scatter(OMC_waist_pos + Closest[0,:],  Closest[1,:]/1e-3, color='b', s=80, marker='x')

        ax2.scatter(OMC_waist_pos + Furthest[0,:], Furthest[2,:]/1e-3, color='r', s=80, marker='+')
        ax2.scatter(OMC_waist_pos + Middle[0,:],   Middle[2,:]/1e-3, color='g', s=80, marker='+')
        ax2.scatter(OMC_waist_pos + Closest[0,:],  Closest[2,:]/1e-3, color='b', s=80, marker='+')

        ax2.scatter(OM1_pos - Middle_SRM_refl_HAM6[0,:], Middle_SRM_refl_HAM6[1,:]/2e-3, color='g', s=80, marker='x')
        ax2.scatter(OM1_pos - Middle_SRM_refl_HAM6[0,:], Middle_SRM_refl_HAM6[2,:]/2e-3, color='g', s=80, marker='+')
        
        ax2.annotate('OMC Waist', xy=(6.15, 0.5), xytext=(6.15, 0.2),
            arrowprops=dict(facecolor='black', arrowstyle='->')
                    )
    
    if lens_config is 'After':
        
        # Double OM3 to OMC path as it travels that kind of 
        # distance again to get to first steering mirror
        OMC_temp = np.copy(OMC_REFL_new)
        OMC1_2_OMC_REFL_ST1 = kat.sOM1_OM2.L + kat.sOM2_OM3.L + kat.sOM3_OMC.L * 2
        OMC_temp[0, :] += OMC1_2_OMC_REFL_ST1
        
        full_HAM5_OMC = np.hstack((HAM5_OMC, OMC_temp))
        
        ax.scatter(ZM1_pos + OPO_new[0,:], OPO_new[1,:]/2e-3, color='g', s=80, marker='x')
        ax.scatter(ZM1_pos + OPO_new[0,:], OPO_new[2,:]/2e-3, color='g', s=80, marker='+')
        
        ax2.scatter(OM1_pos+full_HAM5_OMC[0,:], full_HAM5_OMC[1,:]/2e-3, color='g', s=80, marker='x')
        ax2.scatter(OM1_pos+full_HAM5_OMC[0,:], full_HAM5_OMC[2,:]/2e-3, color='g', s=80, marker='+')
        
        sel = slice(None,None, None)
        zdata = full_HAM5_OMC[0,sel]
        xdata = full_HAM5_OMC[1,sel]/2
        ydata = full_HAM5_OMC[2,sel]/2
        xsig  = full_HAM5_OMC[3,sel]/2
        ysig  = full_HAM5_OMC[0,sel]/2
        
        qSRM_design = pykat.BeamParam(z=-3.55, w0=680e-6)

        kat1 = base.deepcopy()
        
        z = np.linspace(-4.0, 4.45, 1000)

        qFITx,_ = fit_input_q(qSRM_design, zdata, xdata, xsig, z_ref="OM1")
        qFITy,_ = fit_input_q(qSRM_design, zdata, ydata, ysig, z_ref="OM1")
        
        ox, oy = (pykat.BeamParam.overlap(OMCqx, qFITx), pykat.BeamParam.overlap(OMCqy, qFITy))
        
        TFITx = kat1.beamTrace(qFITx, 'nSRMs1', 'nOMC_ICa')
        TFITy = kat1.beamTrace(qFITy, 'nSRMs1', 'nOMC_ICa')
        
        ax2.plot(z+OM1_pos, TFITx.beamsize(z, "OM1")/1e-3,label="Fitted Data (x)")
        ax2.plot(z+OM1_pos, TFITy.beamsize(z, "OM1")/1e-3,label="Fitted Data (y)")
        
        ax2.annotate('OMC Waist', xy=(6.15, 0.5), xytext=(6.15, 0.2),
            arrowprops=dict(facecolor='black', arrowstyle='->')
            )
        
        print("The overlap between (x/y) and OMC is "+
             str(ox) + '/'+str(oy) )
        
        
    ax.grid()
    ax.set_xlabel("OPO Coupler to SRM HR [m]",fontsize=font_labels)
    ax.set_ylabel('Beam Radius [mm]',fontsize=font_labels)
    ax.legend()
    
    ax.set_xlim(0,5.5)
    ax.set_ylim(0,2)
    
    ax2.grid()
    ax2.set_xlabel("SRM HR -> OMC REFL Port [m]",fontsize=font_labels)
    ax2.set_ylabel('Beam Radius [mm]',fontsize=font_labels)
    ax2.legend(ncol=2)
    ax2.set_ylim(0,2.5)

    ax2.set_xlim(0, 8.0)
    
    plt.tight_layout()
    
    if fname is None:
        print('Give me a filename!')
    else:
        plt.savefig('../../Figures/'+fname)
```
