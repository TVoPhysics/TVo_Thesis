```python
%matplotlib inline
import pykat
import pylab as pl
import numpy as np

### My functions
from kat_commands import *
from cav_commands import *
from MMT import modematch

import get_WS
import trace_mode
```

```python
katfile = "aLIGO_OnlySqz_OMC.kat"
maxtem = 0
```

```python
#load the original kat file (ie Perfect Modematching) as a reference
basekat = pykat.finesse.kat() 
basekat.loadKatFile(katfile)
basekat.parseCommands(darm_commands)
basekat.parseCommands("maxtem "+str(maxtem))
basekat.parseCommands("""

bp q_SRM_x x q nSRMARa
bp q_SRM_y y q nSRMARa

bp z_BS_x x z nPRBS
bp z_BS_y y z nPRBS

bp zr_BS_x x zr nPRBS
bp zr_BS_y y zr nPRBS

bp w_BS_x x w nPRBS
bp w_BS_y y w nPRBS

bp r_BS_x x r nPRBS
bp r_BS_y y r nPRBS

bp z_sqz_x x z nOPO_HRIC_in
bp z_sqz_y y z nOPO_HRIC_in

bp w0_sqz_x x w0 nOPO_HRIC_in
bp w0_sqz_y y w0 nOPO_HRIC_in
""")
basekat.undumpNodes() #undump the vacuum nodes
```

```python
#Get the mode of the OMC
kat1 = basekat.deepcopy()
kat1.parseCommands("""
cav cavOMC OMC_HROC nOMC_HROC_refl OMC_HROC nOMC_HROC_in
yaxis re:im
""")
outOMC = kat1.run()

q_OMC = np.complex(outOMC['z_BS_x'],outOMC['zr_BS_x'])
```

```python
outOMC['q_SRM_x']

q_SRM_lisa =  np.complex(3.55,1.09)
```

```python
pykat.BeamParam.overlap(outOMC['q_SRM_x'],q_SRM_lisa)
```

```python
#Set the mode of the Sqz
kat=basekat.deepcopy() #copy kat code
kat.parseCommands("""
gauss gaussOPOx OPO_HRIC nOPO_HRIC_refl 214.0u -0.0592 214.0u -0.0592
""")

kat.parseCommands("""
xaxis* ssqz_lens1 L lin 0.01 -0.01 10
x2axis* ssqz_lens1_lens2 L lin -0.01 0.01 10
""")
out_space_lens1 = kat.run()
```

```python
q_out_x_list = []
q_out_y_list = []
for i in range(0,len(out_space_lens1.x)):
    qx = np.complex(out_space_lens1['z_BS_x'][i][i],out_space_lens1['zr_BS_x'][i][i])
    qy = np.complex(out_space_lens1['z_BS_y'][i][i],out_space_lens1['zr_BS_y'][i][i])
    q_out_x_list.append(qx)
    q_out_y_list.append(qy) 
    
[wx,wy,sx,sy] = get_WS.phase_space(q_out_x_list,q_out_y_list)
```

```python
pl.figure(figsize=(9, 5), dpi=80)
pl.subplot(111)

pl.plot(wx,sx, '--',linewidth=3)

pl.xlabel("Beam Size(mm)")
pl.ylabel("Curvature (microdiopters)")

###Main plot formatting
pl.tick_params(labelsize=12)
pl.tick_params(which='both', width=2)
pl.tick_params(which='major', length=7)
pl.tick_params(which='minor', length=4)
pl.grid(True, zorder=-1)
pl.legend(fontsize=20)

pl.show()
```

```python
deltax = 2
deltay = .2
x = np.arange(-100, 100, deltax)
y = np.arange(-10, 10, deltay)
z_list = []
zr_list = []

for i in range(len(x)):
    z_list.append(q_OMC+x[i])
    zr_list.append(q_OMC+x[i]*1j)

X, Y = np.meshgrid(z_list, zr_list)

Z = pykat.BeamParam.overlap(X,Y)


#### Plot
pl.figure(figsize=(18, 10), dpi=80)

CS = pl.contour(np.real(X),np.imag(Y),Z*100, levels=range(90,101,1))
pl.plot(np.real(q_out_x_list),np.imag(q_out_x_list), '--',linewidth=4)
pl.plot(np.real(q_OMC),np.imag(q_OMC), '*',linewidth=7)

pl.clabel(CS, inline=1, fontsize=20)

pl.xlabel("Re(q)",fontsize=16)
pl.ylabel("Im(q)",fontsize=16)

pl.show()
```
