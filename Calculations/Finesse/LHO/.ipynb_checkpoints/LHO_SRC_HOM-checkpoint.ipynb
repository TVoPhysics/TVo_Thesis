```python
d_SR3 = np.linspace(-0.0007,0.0007,20)
print(d_SR3)
```

```python
results = []
for i in range(len(d_SR3)):
    kat1 = base0.deepcopy()
    #rc_new = (1./kat1.SR3.Rc.value + 1./d_SR3[i])**(-1.)
    kat1.SR3.Rc = kat1.SR3.Rc*(1+d_SR3[i])
    
    print(kat1.SR3.Rc)
    
    kat1.parse('''
    noxaxis
    ''')
    out1 = kat1.run()
    results.append(out1)
```

```python
d_SR3*100
```

```python
fseps = []
gouys = []
for i in range(len(results)):
    fseps.append(results[i]['cavSRX_x_fsep'])
    gouys.append(results[i]['cavSRX_x_gouy'])
    
fseps = np.array(fseps)
gouys = np.array(gouys)
print(fseps)
print(gouys)
```

```python
# Idiot check the transverse mode spacing matches the round trip Gouy phase
(fseps[0]*2*np.pi/srx_fsr)*360/2/np.pi
```

```python
fig = plt.figure(figsize=(5,3),dpi=120)

ax = fig.add_subplot(111)
ax.plot(d_SR3*100,gouys, linewidth=3)

ax.set_xlabel('$\Delta R_{SR3}$ %',fontsize=16)
ax.set_ylabel('Gouy Phase [Deg]',fontsize=16)
ax.tick_params(labelsize=12)

ax2 = ax.twinx()
ax2.plot(d_SR3*100,fseps/1e6, linewidth=3)
ax2.set_ylabel('TMS [MHz]',fontsize=16)
ax2.tick_params(labelsize=12)

plt.grid()
#ax.legend()
```
