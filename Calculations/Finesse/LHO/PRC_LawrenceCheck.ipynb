```python
import pykat
import numpy as np

import pykat.ifo.aligo as aligo
import pykat.ifo.aligo.plot
import matplotlib.pyplot as plt

import scipy
from scipy import signal

pykat.init_pykat_plotting(fmts=['svg'])
```

```python
def P_sb(rp,scatter):
    p_sb = np.sqrt(scatter)/( (1-rp)**2 +rp**2*scatter)
    
    return p_sb
```

```python
scatt = np.linspace(0,1e-3,100)

rp = np.sqrt(0.97)

p_sb_out = P_sb(rp,scatt)
```

```python
fig = plt.figure(figsize=(10,6))

ax = fig.add_subplot(111)

ax.plot(scatt,p_sb_out/max(p_sb_out))

plt.legend()
plt.show()
```
