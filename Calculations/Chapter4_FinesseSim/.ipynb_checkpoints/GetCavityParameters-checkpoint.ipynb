```python
def try_it(kat_in):
    result = None
    tries = None
    while result is None and tries < 20:    
        try:
            out = kat_in.run()
            result = 1
        except:
            tries += 1 
            print("Ran into a pipe error, running again")
            continue

    return out
```

```python
out = basekat.run()
```

```python
#cav_names = ['IMC', 'XARM', 'YARM', 'PRX', 'PRY', 'SRX', 'SRY']
#cavs = ['cavIMC', 'cavXARM', 'cavYARM', 'cavPRX', 'cavPRY', 'cavSRX', 'cavSRY']

cav_names = ['IMC', 'XARM', 'YARM', 'PRX', 'PRY', 'SRX', 'SRY','OMC']
cavs = ['cavIMC', 'cavXARM', 'cavYARM', 'cavPRX', 'cavPRY', 'cavSRX', 'cavSRY','cavOMC']

params = ['zr','z','length','fsr','gouy','pole','finesse','A','B','C','D']
params_name = ['Rayleigh Range [m]','Waist to First Mirror [m]','Cavity Length [m]','FSR [MHz]','RT Gouy Phase [deg]','Pole Frequency [kHz]','Finesse','A','B','C','D']
```

```python
matrix = pandas.DataFrame(columns=cav_names, index=params_name)
```

```python
for i in range(len(cavs)):
    for j in range(len(params)):
        string = cavs[i]+'_x_'+params[j]
        if params[j] is 'fsr':
            matrix[cav_names[i]][params_name[j]] =np.round(out[string],3)/1e6
        elif params[j] is 'pole':
            matrix[cav_names[i]][params_name[j]] =np.round(out[string],3)/1e3
        elif params[j] is 'length':
            matrix[cav_names[i]][params_name[j]] =np.round(out[string],3)/2
        else:
            matrix[cav_names[i]][params_name[j]] =np.round(out[string],3)
```
