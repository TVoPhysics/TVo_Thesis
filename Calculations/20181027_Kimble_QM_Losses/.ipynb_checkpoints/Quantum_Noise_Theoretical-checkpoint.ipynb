```python
import numpy as np
import matplotlib.pyplot as pl

%matplotlib inline
```

```python
#Parameters

#IFO
L = 3995
T = 0.014
m = 40
loss_rt = 150e-6 #arm rt loss

#Power
Pin = 429.8967222619745

#constants
c = 3.e8
hbar = 1.05457e-34
lam = 1064e-9
w0 = (2 * np.pi * c) / lam

#freq
f_sm = np.arange(1e-1, 10,0.1)
f_prmi = np.arange(10, 1e4,0.1)
f_pole = (T*c)/ (4*L)

# Losses
c_loss = (2 * loss_rt) / T #arm loss coeff
ot_loss = 0.05  #fractional photon loss in optical train
arm_loss = (2*c_loss) / (1 + (f/f_pole)**2 ) # fractional photon loss in output train
```

```python
#simple MIchelson
h_sql_sm = np.sqrt( (4.*hbar) / (m * L**2))

h_sql_sm_f = np.sqrt( (4.*hbar) / (m * f**2 * L**2))

kappa_sm = (4*Pin * w0 )/ (m * c**2 * f**2)

shot_noise_sm = (h_sql_sm_f**2 / 2.)*(1./kappa_sm)

rad_pres_sm = (h_sql_sm_f**2 / 2.)* (kappa_sm)

#s_psd_sm = (h_sql_sm_f**2 / 2.)*( (kappa_sm) +(1./kappa_sm) )

s_psd_sm = shot_noise_sm + rad_pres_sm
```

```python
Psql = (m * L**2 * f_pole**4 ) / (4*w0)

kappa  = ( 2 * (Pin/Psql) * f_pole**4 ) / (f_prmi**2 * ( f_pole**2 + f_prmi **2))

h_sql = np.sqrt(8.*hbar / (m * f_prmi**2 * L**2))

rad_prs_prmi = (h_sql**2/2.) * kappa

sht_nse_prmi = (h_sql**2/2.) * (1./kappa)

h_asd =  np.sqrt( sht_nse_prmi + rad_prs_prmi)

test = (2e-24 * f_pole) / (f_prmi)
```

```python
fig = pl.figure(figsize = (12,8),dpi=200)

ax = fig.add_subplot(111)

ax.loglog(f_prmi,h_sql,linewidth=2, label = 'SQL') 
ax.loglog(f_prmi,h_asd,linewidth=2, label = 'Quantum Noise')
ax.loglog(f_prmi,np.sqrt(rad_prs_prmi), '--',label='Radiation Pressure')
ax.loglog(f_prmi,np.sqrt(sht_nse_prmi), '--',label='Shot Noise')

ax.legend()
ax.set_xlim(10, 1e3)
ax.set_ylim(5e-25, 1e-22)

ax.set_xlabel('Frequency [Hz]', fontsize = 20)
ax.set_ylabel('Strain $1/\sqrt{Hz}$',fontsize =20)
ax.tick_params(labelsize=16)

ax.grid()
```
